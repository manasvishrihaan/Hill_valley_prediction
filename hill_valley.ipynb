Hill and Valley Prediction

objective:Given data set where we have to determine the hill and valley prediction.Each record has 100 points on a 2D graph which determines if its a hill or a valley by plotting them on a graph
y='Class',x='v1,v2,v3.......................v100'.

Import Libraury

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import data

```python
hills=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Hill%20Valley%20Dataset.csv')
```

Describe Data

```python
hills.head()
```

No of rows and columns

```python
hills.shape
```

```python
hills.info()
```

```python
hills.describe()
```

```python
hills.columns
```

Define target variable(y)

```python
#dependent variable
y=hills['Class']
```

Define feature variables(x)

```python
#independent varaibles
x=hills.drop('Class',axis=1)
```

Plot the graph

```python
plt.plot(x,y)
plt.xlabel('Points')
plt.ylabel('Class')
```

```python
plt.plot(x.iloc[1,:])
```

Train_Test_Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529,stratify=y)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

Modeling

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train,y_train)
```

Prediction

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

Final Result

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

Finally we got the accuracy of 96 percent by using the LogisticRegression model.Since the target variable were categorical we used the LogisticRegression method

```python
classification_report(y_test,y_pred)
```
